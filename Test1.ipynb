```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
```

```python
# Load dataset
df = pd.read_csv('creditcard.csv')

print('Jumlah baris: ' + str(df.shape[0]))
print('Jumlah kolom: ' + str(df.shape[1]))
```

```python
# Cek distribusi kelas
class_counts = df['Class'].value_counts()
class_pct = df['Class'].value_counts(normalize=True) * 100

print('================================')
print('DISTRIBUSI KELAS TRANSAKSI')
print('================================')
print('Normal (0) :', class_counts[0], 'transaksi')
print('Fraud  (1) :', class_counts[1], 'transaksi')
print('================================')
print('Persentase Normal :', round(class_pct[0], 4), '%')
print('Persentase Fraud  :', round(class_pct[1], 4), '%')
print('================================')
```

```python
print('Missing Vlue per kolom:')
print(df.isnull().sum())
```

```python
duplcates = df.duplicated().sum()
print ('Jumlah data duplikat:', duplcates)

if duplcates > 0:
    df.drop_duplicates(inplace=True)
    print(str(duplcates) + 'baris duplikat berhasil di hapus')
else:
    print('Tidak ada data duplikat')
```

```python
# Visualisasi distribusi kelas
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
colors = ['#2ecc71', '#e74c3c']
axes[0].bar(['Normal (0)', 'Fraud (1)'], class_counts.values, 
            color=colors, edgecolor='black', width=0.5)
axes[0].set_title('Distribusi Kelas Transaksi', fontsize=14)
axes[0].set_ylabel('Jumlah Transaksi')

# Pie chart
axes[1].pie(class_counts.values, 
            labels=['Normal', 'Fraud'], 
            colors=colors,
            autopct='%1.3f%%', 
            startangle=90, 
            explode=(0, 0.1))
axes[1].set_title('Proporsi Kelas Transaksi', fontsize=14)

plt.suptitle('Class Imbalance Analysis', fontsize=16)
plt.tight_layout()
plt.show()

print('Insight: Dataset sangat tidak seimbang (imbalanced).')
print('Hanya 0.17% dari total transaksi merupakan fraud.')
print('Kondisi ini mencerminkan situasi nyata di industri pembayaran.')
```

```python
# Distribusi nilai transaksi (Amount)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram semua transaksi
axes[0].hist(df['Amount'], bins=100, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].set_title('Distribusi Nilai Transaksi (Amount)', fontsize=13)
axes[0].set_xlabel('Amount (Euro)')
axes[0].set_ylabel('Frekuensi')
axes[0].axvline(df['Amount'].median(), color='red', linestyle='--', 
                label='Median: ' + str(round(df['Amount'].median(), 2)))
axes[0].legend()

# Boxplot Normal vs Fraud
df.boxplot(column='Amount', by='Class', ax=axes[1])
axes[1].set_title('Amount: Normal vs Fraud', fontsize=13)
axes[1].set_xlabel('Class (0=Normal, 1=Fraud)')
axes[1].set_ylabel('Amount (Euro)')
plt.suptitle('')

plt.tight_layout()
plt.show()

normal_median = df[df['Class']==0]['Amount'].median()
fraud_median = df[df['Class']==1]['Amount'].median()
print('Median Amount Normal :', round(normal_median, 2))
print('Median Amount Fraud  :', round(fraud_median, 2))
print('Insight: Transaksi fraud cenderung bernilai lebih rendah dari transaksi normal.')
```

```python
# Konversi kolom Time (detik) ke jam
df['Hour'] = (df['Time'] / 3600).astype(int) % 24

# Hitung jumlah transaksi per jam per kelas
normal_hour = df[df['Class']==0]['Hour'].value_counts().sort_index()
fraud_hour = df[df['Class']==1]['Hour'].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Transaksi normal per jam
axes[0].plot(normal_hour.index, normal_hour.values, 
             color='#2ecc71', linewidth=2)
axes[0].set_title('Pola Transaksi Normal per Jam', fontsize=13)
axes[0].set_xlabel('Jam dalam Sehari')
axes[0].set_ylabel('Jumlah Transaksi')

# Transaksi fraud per jam
axes[1].bar(fraud_hour.index, fraud_hour.values, 
            color='#e74c3c', alpha=0.8)
axes[1].set_title('Pola Transaksi Fraud per Jam', fontsize=13)
axes[1].set_xlabel('Jam dalam Sehari')
axes[1].set_ylabel('Jumlah Transaksi Fraud')

plt.suptitle('Analisis Pola Waktu Transaksi', fontsize=15)
plt.tight_layout()
plt.show()

peak_fraud_hour = fraud_hour.idxmax()
print('Jam paling banyak terjadi fraud:', str(peak_fraud_hour) + ':00')
print('Insight: Fraud lebih sering terjadi saat volume transaksi normal rendah.')
print('Ini membuat fraud lebih sulit terdeteksi secara manual.')
```

```python
# Hitung selisih mean antara normal dan fraud untuk tiap fitur V
v_features = ['V' + str(i) for i in range(1, 29)]

mean_normal = df[df['Class']==0][v_features].mean()
mean_fraud = df[df['Class']==1][v_features].mean()
mean_diff = abs(mean_normal - mean_fraud).sort_values(ascending=False)

plt.figure(figsize=(14, 5))
colors = ['#e74c3c' if v in mean_diff.head(5).index else 'steelblue' 
          for v in mean_diff.index]
plt.bar(mean_diff.index, mean_diff.values, color=colors, edgecolor='white')
plt.title('Selisih Mean Fitur V: Normal vs Fraud', fontsize=14)
plt.xlabel('Fitur')
plt.ylabel('Selisih Mean')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print('Top 5 Fitur Paling Membedakan Fraud vs Normal:')
for i, (feat, val) in enumerate(mean_diff.head(5).items(), 1):
    print(str(i) + '. ' + feat + ' : ' + str(round(val, 4)))
```

```python
# Distribusi top 6 fitur V: Normal vs Fraud
top6_features = mean_diff.head(6).index.tolist()

fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

for i, feat in enumerate(top6_features):
    df[df['Class']==0][feat].hist(ax=axes[i], bins=50, 
                                   color='#2ecc71', alpha=0.6, 
                                   label='Normal', density=True)
    df[df['Class']==1][feat].hist(ax=axes[i], bins=50, 
                                   color='#e74c3c', alpha=0.6, 
                                   label='Fraud', density=True)
    axes[i].set_title('Distribusi ' + feat, fontsize=12)
    axes[i].legend(fontsize=9)
    axes[i].set_xlabel(feat)
    axes[i].set_ylabel('Density')

plt.suptitle('Distribusi Top 6 Fitur: Normal vs Fraud', fontsize=14)
plt.tight_layout()
plt.show()

print('Insight: Fitur-fitur di atas menunjukkan distribusi yang')
print('sangat berbeda antara transaksi normal dan fraud.')
print('Ini menjadikannya kandidat fitur terkuat untuk model ML selanjutnya.')
```

```python
# Korelasi fitur terhadap Class
corr_with_class = df.corr()['Class'].drop('Class').sort_values(key=abs, ascending=False)

plt.figure(figsize=(10, 7))
colors = ['#e74c3c' if x < 0 else '#2ecc71' for x in corr_with_class.values]
plt.barh(corr_with_class.index, corr_with_class.values, 
         color=colors, edgecolor='white')
plt.axvline(0, color='black', linewidth=0.8)
plt.title('Korelasi Fitur terhadap Class (Fraud)', fontsize=14)
plt.xlabel('Korelasi')
plt.tight_layout()
plt.show()

print('Korelasi POSITIF tertinggi terhadap fraud:')
print(corr_with_class[corr_with_class > 0].head(5))
print()
print('Korelasi NEGATIF terkuat terhadap fraud:')
print(corr_with_class[corr_with_class < 0].head(5))
```

```python
# Heatmap korelasi antar top fitur
top_features = corr_with_class.abs().head(10).index.tolist()
top_features = top_features + ['Amount', 'Class']
top_features = list(dict.fromkeys(top_features))

plt.figure(figsize=(12, 8))
corr_matrix = df[top_features].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm',
            vmin=-1, 
            vmax=1, 
            mask=mask, 
            linewidths=0.5)
plt.title('Heatmap Korelasi — Top Fitur vs Class', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Summary statistik akhir
print('=' * 55)
print('RINGKASAN ANALISIS FRAUD DETECTION EDA')
print('=' * 55)
print('Total transaksi      :', len(df))
print('Transaksi normal     :', class_counts[0])
print('Transaksi fraud      :', class_counts[1])
print('Persentase fraud     :', round(class_pct[1], 4), '%')
print('Max Amount Normal    :', round(df[df['Class']==0]['Amount'].max(), 2))
print('Max Amount Fraud     :', round(df[df['Class']==1]['Amount'].max(), 2))
print('Median Amount Normal :', round(df[df['Class']==0]['Amount'].median(), 2))
print('Median Amount Fraud  :', round(df[df['Class']==1]['Amount'].median(), 2))
print('Top fitur signifikan :', ', '.join(mean_diff.head(5).index.tolist()))
print('=' * 55)
```

Insight

### Temuan Utama
1. Dataset sangat imbalanced — hanya 0.17% transaksi adalah fraud
2. Fraud lebih sering terjadi pada jam dini hari saat volume transaksi rendah
3. Nilai transaksi fraud cenderung lebih kecil dari transaksi normal
4. Fitur V14, V17, V12, V10, V16 paling membedakan fraud vs normal

### Rekomendasi Bisnis
1. Terapkan monitoring intensif pada jam 00.00–06.00
2. Jangan hanya andalkan nilai Amount sebagai trigger alert fraud
3. Gunakan fitur V14 & V17 sebagai indikator awal fraud
4. Gunakan SMOTE untuk mengatasi class imbalance sebelum modeling
5. Evaluasi model dengan AUC-ROC dan Precision-Recall, bukan hanya akurasi
